```python
# 直接用原始数据作图（时间 vs CO2 capacity）
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("=" * 80)
print("CO2 Capacity 时间序列图（原始数据）")
print("=" * 80)

# 读取 Excel 文件
try:
    df = pd.read_excel('day1.xlsx', engine='openpyxl')
    print(f"\n✓ 成功读取 Excel 文件")
except Exception as e:
    raise RuntimeError(f"读取 Excel 文件失败：{e}")

# 获取列信息
first_col = df.columns[0]
last_col = df.columns[-1]

print(f"\n数据信息：")
print(f"  • 第一列（时间）：'{first_col}'")
print(f"  • 最后一列（CO2）：'{last_col}'")
print(f"  • 总行数：{len(df)}")

# 提取原始数据
x_data = df.iloc[:, 0]      # 第一列（时间）
y_data = df.iloc[:, -1]     # 最后一列（CO2）

print(f"\n数据统计：")
print(f"  • X 轴数据点数：{len(x_data)}")
print(f"  • Y 轴数据点数：{len(y_data)}")

# 尝试解析时间列
x_parsed = pd.to_datetime(x_data, errors='coerce')
x_valid = x_parsed.notna().sum()
print(f"  • 时间列解析成功：{x_valid}/{len(x_data)} 行")

# 转换 Y 轴为数值（用于绘图）
y_numeric = pd.to_numeric(y_data, errors='coerce')
y_valid = y_numeric.notna().sum()

print(f"  • CO2 值解析成功：{y_valid}/{len(y_data)} 行")
print(f"  • CO2 值范围：{y_numeric.min():.6f} ~ {y_numeric.max():.6f}")

# 提取小时信息（从时间中只取小时部分）
hours = x_parsed.dt.strftime('%H:%M').fillna('NaN')

# 绘图
fig, ax = plt.subplots(figsize=(16, 7))

# 直接绘制原始数据
ax.plot(range(len(x_data)), y_numeric, 
        marker='o', linestyle='-', linewidth=1.5, markersize=3, 
        color='steelblue', alpha=0.8, label='CO2 Capacity')

# 设置标签和标题
ax.set_xlabel('时间 (小时:分钟)', fontsize=13, fontweight='bold')
ax.set_ylabel('CO2 Capacity', fontsize=13, fontweight='bold')
ax.set_title(f'CO2 Capacity vs 时间 (共 {len(x_data)} 行数据)', 
             fontsize=15, fontweight='bold', pad=20)

# 网格
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='best')

# 设置 X 轴标签为时间（每隔一定间隔显示一个）
tick_interval = max(1, len(x_data) // 20)  # 大约显示 20 个时间点
x_positions = range(0, len(x_data), tick_interval)
x_labels = [hours.iloc[i] if i < len(hours) else 'NaN' for i in x_positions]
ax.set_xticks(x_positions)
ax.set_xticklabels(x_labels, rotation=45, ha='right')

plt.tight_layout()
plt.show()

# 保存图片
try:
    plt.savefig('co2_capacity_timeseries_raw.png', dpi=150, bbox_inches='tight')
    print("\n✓ 图表已保存为 'co2_capacity_timeseries_raw.png'")
except Exception as e:
    print(f"\n⚠ 保存图表失败：{e}")

# 显示原始数据样本
print(f"\n原始数据样本（前 20 行）：")
print("-" * 80)
print(f"{'行号':<5} | {'时间（小时:分钟）':<20} | {'CO2':<15}")
print("-" * 80)
for i in range(min(20, len(df))):
    time_str = hours.iloc[i]
    co2_val = str(y_data.iloc[i])
    print(f"{i+1:<5} | {time_str:<20} | {co2_val:<15}")

print("-" * 80)
print("\n" + "=" * 80)
```
